```python
#Controlling Axes and Adding Diagonals
g = sns.FacetGrid(movies,row='Genre',col='Year',hue='Genre')
kws = dict(s=50 , linewidth=0.5 , edgecolor='black')
g = g.map(plt.scatter,'CriticRatings','AudienceRatings',**kws)
g.set(xlim=(0,100),ylim=(0,100))
for i in g.axes.flat:
    i.plot((0,100),(0,100), c='gray',ls='--')
g.add_legend()
```

```python
#Thematic Edits
```

```python
sns.set_style('whitegrid')
list1 = list()
mylabels = list()
for i in movies.Genre.unique():
    list1.append(movies[movies.Genre==i].Budget)
    mylabels.append(i)
fig,ax=plt.subplots()    
h = plt.hist(list1,bins=30,stacked=True,label=mylabels)
fig.set_size_inches(11.7,8,27)
plt.title('Movie Budget Distribution',fontsize=35)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(frameon=True,fancybox=True,shadow=True,prop={'size':15})
plt.show()
```
